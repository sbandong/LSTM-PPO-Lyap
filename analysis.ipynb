```python
import numpy as np
import matplotlib.pyplot as plt

# time setting
t_end = 15
dt = 0.02
t = np.arange(0, t_end + dt, dt)

d0 = np.zeros_like(t)  # No disturbance

# HasilWithDisturbance1
d1 = np.zeros_like(t)
d1[t <= 2] = 0.2
d1[(t >= 2) & (t <= 3)] = -0.2

# HasilWithDisturbance2
d2 = np.zeros_like(t)
d2[t <= 1] = -0.3
d2[(t >= 3) & (t <= 4)] = 0.1

# HasilWithDisturbance3
d3 = np.zeros_like(t)

# Corrected assumption: 1.2s to 1.3s
d3[(t >= 1.2) & (t <= 1.3)] = -0.15
d3[(t >= 4) & (t <= 4.3)] = 0.25

# Plot
plt.figure(figsize=(10, 4))
plt.plot(t, d0, label='No Disturbance')
plt.plot(t, d1, label='Disturbance 1')
plt.plot(t, d2, label='Disturbance 2')
plt.plot(t, d3, label='Disturbance 3')

plt.xlabel('Time (s)')
plt.ylabel('Disturbance')
# plt.title('Disturbance Profiles')
plt.xlim(0, 15)
plt.grid(True)
plt.legend()
plt.show()
```
